```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Load CSV files
train_csv = pd.read_csv('/content/sign_mnist_train.zip')
test_csv = pd.read_csv('/content/sign_mnist_test.zip')

# Assuming images are stored in a folder and image paths in CSV are relative
train_image_dir = 'D:\Sign_language_Detection_notRLT\sign_mnist_train.zip'
test_image_dir = 'D:\Sign_language_Detection_notRLT\sign_mnist_test.zip'

# Function to load images from file paths and labels
def load_data_from_csv(csv_file, img_dir):
    images = []
    labels = []
    for index, row in csv_file.iterrows():
        img_path = os.path.join(img_dir, row['image'])
        image = cv2.imread(img_path)
        image = cv2.resize(image, (64, 64))
        images.append(image)
        labels.append(row['label'])
    return np.array(images), np.array(labels)

# Load train and test data
X_train, y_train = load_data_from_csv(train_csv, train_image_dir)
X_test, y_test = load_data_from_csv(test_csv, test_image_dir)

# Preprocessing: Normalize image data
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Convert labels to categorical (one-hot encoding)
class_labels = sorted(train_csv['label'].unique())  # Ensure labels are ordered alphabetically
label_map = {label: i for i, label in enumerate(class_labels)}
y_train = to_categorical([label_map[label] for label in y_train], num_classes=len(class_labels))
y_test = to_categorical([label_map[label] for label in y_test], num_classes=len(class_labels))

# Data augmentation
train_datagen = ImageDataGenerator(rotation_range=20, zoom_range=0.2, shear_range=0.2, horizontal_flip=True)
train_datagen.fit(X_train)

# CNN Model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),

    Dense(len(class_labels), activation='softmax')
])

# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train model
history = model.fit(train_datagen.flow(X_train, y_train, batch_size=32), epochs=20, validation_data=(X_test, y_test))

# Save model
model.save('sign_language_model.h5')

# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title('Accuracy')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title('Loss')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical

# Load CSV files
train_csv = pd.read_csv('/content/sign_mnist_train.zip')
test_csv = pd.read_csv('/content/sign_mnist_test.zip')

# Separate labels and pixel values
X_train = train_csv.drop(columns=['label']).values  # pixel values
y_train = train_csv['label'].values  # labels

X_test = test_csv.drop(columns=['label']).values  # pixel values
y_test = test_csv['label'].values  # labels

# Reshape pixel data into 28x28 images (assuming images are grayscale, as we have 784 pixels)
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

# Normalize the pixel values to [0, 1]
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Convert to pandas series to use fillna, if necessary
y_train = pd.Series(y_train).apply(pd.to_numeric, errors='coerce').fillna(0).astype(int).values
y_test = pd.Series(y_test).apply(pd.to_numeric, errors='coerce').fillna(0).astype(int).values

# Convert labels to categorical (one-hot encoding)
num_classes = len(np.unique(y_train))  # Number of unique classes in the labels
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)

# CNN Model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),

    Dense(num_classes, activation='softmax')
])

# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, batch_size=32, epochs=20, validation_data=(X_test, y_test))

# Save the trained model
model.save('sign_language_model.h5')

# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title('Accuracy')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title('Loss')
plt.show()
```
